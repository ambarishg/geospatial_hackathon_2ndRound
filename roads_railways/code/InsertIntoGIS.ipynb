```python
import geopandas as gpd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# create the sqlalchemy connection engine
db_connection_url = "postgresql://ambarish:%s@ambarishpgsql.postgres.database.azure.com/socioeconomic" % quote_plus("Kolkata@1234")
con = create_engine(db_connection_url)
```

```python
FILE1 = '../input/roads_railways/gis_osm_roads_free_1.SHP'
table_name = "roads"
```

```python
%%time
gdf = gpd.read_file(FILE1)
# Drop nulls in the geometry column
print('Dropping ' + str(gdf.geometry.isna().sum()) + ' nulls.')
gdf = gdf.dropna(subset=['geometry'])
```

```python
len(gdf)/1e6
```

```python
# Push the geodataframe to postgresql
gdf.to_postgis(table_name, con, index=False, if_exists='replace') 
```

```python
q = "SELECT * FROM " + table_name + " WHERE ac_name" +  " = '" + p  + "';"
```

```python
%%time
df = gpd.GeoDataFrame.from_postgis(q,
 con,
 geom_col='geometry')
```
